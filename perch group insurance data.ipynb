### c. Is the proportion of smokers significantly different in different genders? 


```python
df1 = df 
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
sex_encoded =  le.fit_transform(df['sex'])
smoker_encoded = le.fit_transform(df['smoker'])
print(sex_encoded)
print(smoker_encoded)
```

```python
df['encoded_sex'] = sex_encoded
df['encoded_smoker'] = smoker_encoded
print(df.head())        # Female = '0' & Male= '1'
```

```python
sns.pairplot(df, hue = 'smoker')
plt.plot()
```

```python
sns.countplot(x = 'encoded_sex', data = df)
```

```python
sns.countplot(x = 'encoded_smoker', data = df)
```

From the pair plot it is clearly visible that there is no proportion of smokers significantly different in different genders.
